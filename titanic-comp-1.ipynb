```python

# Extract the target variable 'Transported' from the 'data' DataFrame
y = data['Transported']

# Define the final features to be used in the model
final_features = ['FoodCourt', 'Age', 'VRDeck', 'Spa', 'RoomService', 'ShoppingMall', 'CryoSleep', 'Destination', 'HomePlanet', 'Cabin']

# Extract the features from the 'data' DataFrame based on the final features list
X = data[final_features]

# Extract the features from the 'test' DataFrame based on the final features list
test = test[final_features]

# Split the data into training and testing sets using the train_test_split function
# The training set (X_train, y_train) will be 75% of the data, and the testing set (X_test, y_test) will be 25%
# The random_state parameter ensures reproducibility of the split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# test = test.drop(columns=['Name','PassengerId'])

```

```python
# Define categorical features in the dataset
categorical_features = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination']

# Define numeric features in the dataset
numeric_features = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

```

```python
print(data.info())
```

```python
# Define a pipeline for processing categorical features
categorical_transformer = Pipeline(steps=[
    # Impute missing values in categorical features with the most frequent category
    ('imputer', SimpleImputer(strategy='most_frequent')),
    
    # One-hot encode categorical features, handling unknown categories by ignoring them
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Define a pipeline for processing numeric features
numeric_transformer = Pipeline(steps=[
    # Scale numeric features, for example, using StandardScaler
    ('scaler', StandardScaler())
])
```

```python
# Define a ColumnTransformer to apply different transformers to different subsets of features
preprocessor = ColumnTransformer(
    transformers=[
        # Apply the categorical_transformer to categorical features
        ('cat', categorical_transformer, categorical_features),
        
        # Apply the numeric_transformer to numeric features
        ('num', numeric_transformer, numeric_features)
    ])

# Create a classification pipeline
clf = Pipeline(steps=[
    # Apply the preprocessor defined by the ColumnTransformer
    ('preprocessor', preprocessor),
    
    # Use a classifier, for example, RandomForestClassifier (replace with your chosen classifier)
    ('classifier', RandomForestClassifier())
])

```
